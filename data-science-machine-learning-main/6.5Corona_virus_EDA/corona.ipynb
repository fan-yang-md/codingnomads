```python
covid['ObservationDate']=pd.to_datetime(covid['ObservationDate'])
covid['Last Update'] = pd.to_datetime(covid['Last Update'])
covid.info()
```

```python
covid.drop(['SNo'], 1, inplace=True)
```

```python
covid.info()
```

```python
covid_by_country = covid.groupby(['Country/Region', 'ObservationDate']).agg({'Confirmed':'sum', 'Recovered':'sum', 'Deaths':'sum'})
covid_by_country.head(20)
```

```python
covid_by_date = covid.groupby(['ObservationDate']).agg({'Confirmed':'sum', 'Recovered':'sum', 'Deaths':'sum'})
```

```python
covid_by_date.head()
```

```python
covid_by_date['Active Cases'] = covid_by_date['Confirmed']-covid_by_date['Recovered']-covid_by_date['Deaths']
fig = px.bar(x=covid_by_date.index, y=covid_by_date['Active Cases'])
fig.show()
covid_by_date.head()
```

```python
fig.show()
```

```python
fig_death = px.line(x=covid_by_date.index, y=covid_by_date['Deaths', 'Active Cases'])
fig_death.show()
```
